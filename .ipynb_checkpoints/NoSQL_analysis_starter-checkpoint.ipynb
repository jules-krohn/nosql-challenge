```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print('The number of restaurants with a hygiene score of 20:', count)
```

```python
# Display the first document in the results using pprint
print("First result: ")
results = establishments.find(query)
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
import pandas as pd 

df = pd.DataFrame(results)
df
```

```python
# Display the number of rows in the DataFrame
print(len(df), "rows in the DataFrame")

# # Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query2 = {'LocalAuthorityName': {'$regex': 'London'},'RatingValue': {'$gte': 4}}
# Use count_documents to display the number of documents in the result

counts = establishments.count_documents(query2)
print("Number of London restaurants with a rating of 4 or above:", counts)
```

```python
# Display the first document in the results using pprint
ratings_results = establishments.find(query2)
pprint(ratings_results[0])
```

```python
# Convert the result to a Pandas DataFrame
ratings_df = pd.DataFrame(ratings_results)
ratings_df
```

```python
# Display the number of rows in the DataFrame
print( len(ratings_df), 'rows in the DataFrame')
# Display the first 10 rows of the DataFrame
ratings_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by lowest hygiene score

#variables
degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

#Query to find the rating and appropriate latitude/longitude measurements
query3 = {'RatingValue': 5, 'geocode.longitude': {'$gte' : (longitude - degree_search), '$lte': (longitude + degree_search)},
        'geocode.latitude': {'$gte': (latitude - degree_search), '$lte': (latitude + degree_search)}}
sort =  [('scores.Hygiene', 1)]

# Assign the query to a variable
rating_hygiene_results = establishments.find(query3).sort(sort).limit(5)

# Pretty print each result
for result in rating_hygiene_results:
    pprint(result)
```

```python
# Convert above results to Pandas DataFrame
df2 = pd.DataFrame(establishments.find(query3).sort(sort).limit(5))
df2
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0

match_query = {'$match': {'scores.Hygiene': 0}}
# 2. Groups the matches by Local Authority

group_query = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}

# 3. Sorts the matches from highest to lowest

sort_values = {'$sort': {'count': -1}}

# Put pipeline together
pipeline = [match_query, group_query, sort_values]

# Run pipeline through the aggregate method and assign to a variable

results4 = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in result:", len(results4))

# Print the first 10 results
pprint(results4[0:10])
```

```python
# Convert the result to a Pandas DataFrame
df4 = pd.DataFrame(results4)

#Rename columns
df4.rename(columns={"_id":"Local Authority", "count":"Number of Establishments"}, inplace=True)
df4
```

```python
# Display the number of rows in the DataFrame
print(len(results4), "rows in this DataFrame")

# Display the first 10 rows of the DataFrame
df4.head(10)
```
